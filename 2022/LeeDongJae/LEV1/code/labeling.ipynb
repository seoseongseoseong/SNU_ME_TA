```python
### Reading Data
import os
from glob import glob
import pandas as pd
#os.listdir('/content/drive/MyDrive/csv')
#forders = os.listdir('/content/drive/MyDrive/csv')
file_names = glob("/content/drive/MyDrive/csv/R002SMDI.csv")
dataframe_raw = pd.DataFrame()

for file_name in file_names:
    temp = pd.read_csv(file_name)
    dataframe_raw = pd.concat([dataframe_raw, temp], ignore_index=True)
```
